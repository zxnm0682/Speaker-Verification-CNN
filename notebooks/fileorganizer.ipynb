```python
import os
import wave
import contextlib
import time
import gc
import stat

def delete_wav_outside_range(root_folder, min_sec=5, max_sec=20):
    for dirpath, _, filenames in os.walk(root_folder):
        for filename in filenames:
            if filename.lower().endswith(".wav"):
                file_path = os.path.join(dirpath, filename)
                try:
                    with contextlib.closing(wave.open(file_path, 'r')) as wav_file:
                        frames = wav_file.getnframes()
                        rate = wav_file.getframerate()
                        duration = frames / float(rate)

                    gc.collect()  #    

                    if duration < min_sec or duration > max_sec:
                        print(f"Deleting: {file_path} (Duration: {duration:.2f} sec)")

                        #   ()
                        if not os.access(file_path, os.W_OK):
                            os.chmod(file_path, stat.S_IWRITE)

                        time.sleep(0.1)  # 
                        os.remove(file_path)

                except wave.Error as e:
                    print(f"Skipping (wave error): {file_path} - {e}")
                except Exception as e:
                    print(f"Skipping (other error): {file_path} - {e}")


#  
#    
root_path = r"C:\workspace\dataset\test"
delete_wav_outside_range(root_path)
```
